```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso

data=pd.read_csv('student_scores - student_scores.csv')
plt.figure()
plt.scatter(data['Hours'],data['Scores'])
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
X_train,X_test,y_train,y_test = train_test_split(np.array(data['Hours']).reshape(-1, 1),np.array(data['Scores']).reshape(-1, 1),train_size=0.8,random_state=0)
```

```
model=LinearRegression().fit(X_train,y_train)
print('Linear Regression score:\n','Training score:',model.score(X_train,y_train),'Testing Score:',model.score(X_test,y_test),'Score for 9.25hrs/day:',model.predict(np.array([9.25]).reshape(-1,1)))
plt.figure()
plt.scatter(data['Hours'],data['Scores'])
y=model.coef_*data['Hours'].values + model.intercept_
plt.plot(data['Hours'],y.reshape(-1,1))
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```
model=Ridge().fit(X_train,y_train)
print('Ridge Regression score:\n','Training score:',model.score(X_train,y_train),'Testing Score:',model.score(X_test,y_test),'Score for 9.25hrs/day:',model.predict(np.array([9.25]).reshape(-1,1)))
plt.figure()
plt.scatter(data['Hours'],data['Scores'])
y=model.coef_*data['Hours'].values + model.intercept_
plt.plot(data['Hours'],y.reshape(-1,1))
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```
model=Lasso().fit(X_train,y_train)
print('Lasso Regression score:\n','Training score:',model.score(X_train,y_train),'Testing Score:',model.score(X_test,y_test),'Score for 9.25hrs/day:',model.predict(np.array([9.25]).reshape(-1,1)))
plt.figure()
plt.scatter(data['Hours'],data['Scores'])
y=model.coef_*data['Hours'].values + model.intercept_
plt.plot(data['Hours'],y.reshape(-1,1))
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```
